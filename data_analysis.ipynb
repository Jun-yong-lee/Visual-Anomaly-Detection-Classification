## Data Analysis

```python
import os
import matplotlib.pyplot as plt

mvtec_imagefolder_path = "mvtec_anomaly_detection_imagefolder"
class_image_counts = {}

for class_name in os.listdir(mvtec_imagefolder_path):
    class_path = os.path.join(mvtec_imagefolder_path, class_name)
    if os.path.isdir(class_path):
        image_count = len([file for file in os.listdir(class_path) if os.path.isfile(os.path.join(class_path, file))])
        class_image_counts[class_name] = image_count

sorted_class_image_counts = {k: v for k, v in sorted(class_image_counts.items(), key=lambda item: item[1])}

print(f"class:counts = \n{sorted_class_image_counts}")
```

```python
print(f"{sum(sorted_class_image_counts.values())}")
```

### Visualization

```python
classes = list(sorted_class_image_counts.keys())
image_counts = list(sorted_class_image_counts.values())

plt.figure(figsize=(24, 6))
plt.bar(classes, image_counts)
plt.xlabel('Class name')
plt.ylabel('Image counts')
plt.title('Class-Imagecounts')
plt.xticks(rotation=45, ha='right')
plt.grid()
plt.show()
```
